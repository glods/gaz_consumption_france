### 1.2. Download gas price data from the GitHub repository

The data includes: **Price files**, **Stations file** and the **Services file** previously define

```python
# ============================================================================
# FUNCTION TO DATA DOWNLOAD 
# ============================================================================

def download_gas_files(data_dir: str, base_url: str, first_year: int, last_year: int) -> dict:
    """
    Download gas price data files from GitHub repository.
    
    Parameters:
    ----------
    data_dir : str
        Directory to save downloaded files
    base_url : str
        Base URL of the GitHub raw content
    first_year : int
        First year to download
    last_year : int
        Last year to download
        
    Returns:
    --------
    dict : Dictionary containing paths to downloaded files organized by type
    """
    # Create data directory
    Path(data_dir).mkdir(parents=True, exist_ok=True)
    
    downloaded_files = {
        'prices': [],
        'stations': None,
        'services': None
    }
    
    # Define files to download
    files_to_download = []
    
    # Price files for each year
    for year in range(first_year, last_year + 1):
        if year == 2022:
            # 2022 is split into two parts
            files_to_download.append((f"Prix{year}S1.csv.gz", 'prices'))
            files_to_download.append((f"Prix{year}S2.csv.gz", 'prices'))
        else:
            files_to_download.append((f"Prix{year}.csv.gz", 'prices'))
    
    # Station and Services files (2024 versions)
    files_to_download.append(("Stations2024.csv.gz", 'stations'))
    files_to_download.append(("Services2024.csv.gz", 'services'))
   
    # Download each file
    for filename, file_type in files_to_download:
        url = f"{base_url}/{filename}"
        local_path = os.path.join(data_dir, filename)
        csv_path = local_path.replace('.gz', '')
        
        try:
            if not os.path.exists(csv_path):
                print(f"Downloading {filename}...", end=" ")
                urllib.request.urlretrieve(url, local_path)
                
                # Decompress .gz file
                with gzip.open(local_path, 'rb') as f_in:
                    with open(csv_path, 'wb') as f_out:
                        shutil.copyfileobj(f_in, f_out)
                
                # Remove compressed file to save space
                os.remove(local_path)
                print("Good")
            else:
                print(f"File {csv_path} already exists, skipping...")
            
            # Track downloaded files
            if file_type == 'prices':
                downloaded_files['prices'].append(csv_path)
            elif file_type == 'stations':
                downloaded_files['stations'] = csv_path
            elif file_type == 'services':
                downloaded_files['services'] = csv_path
                
        except Exception as e:
            print(f"Bad :  Error: {e}")
    
    return downloaded_files
```

```python
# Download all required files
print("=" * 60)
print("DOWNLOADING DATA FILES")
print("=" * 60)

try:
    downloaded = download_gas_files(
        data_dir=config['data_directory'],
        base_url=config['github_base_url'],
        first_year=config['first_year'],
        last_year=config['last_year']
    )
    print("\n" + "=" * 60)
    print("Downloaded files summary:")
    print(f" - Price files: {len(downloaded['prices'])}")
    print(f" - Stations file: {downloaded['stations']}")
    print(f" - Services file: {downloaded['services']}")

except Exception as e:
    print(f"\n⚠ Automatic download failed: {e}")
    print("\n" + "=" * 60)
    print("MANUAL DOWNLOAD INSTRUCTIONS")
    print("=" * 60)
    print("""
Please manually download the following files from:
https://github.com/rvm-courses/GasPrices

Required files:
  1. Prix2022S1.csv.gz -> decompress to Prix2022S1.csv
  2. Prix2022S2.csv.gz -> decompress to Prix2022S2.csv
  3. Prix2023.csv.gz   -> decompress to Prix2023.csv
  4. Prix2024.csv.gz   -> decompress to Prix2024.csv
  5. Stations2024.csv.gz -> decompress to Stations2024.csv
  6. Services2024.csv.gz -> decompress to Services2024.csv

Steps:
  1. Go to the GitHub repo and click on each file
  2. Click 'Download raw file' button
  3. Decompress .gz files using: gunzip <filename>.gz
  4. Place all .csv files in the ./data/ directory
  5. Re-run this cell
""")
    
    # Check if files already exist (manual download)
    data_dir = config['data_directory']
    price_files = []
    for year in range(config['first_year'], config['last_year'] + 1):
        if year == 2022:
            price_files.append(f"{data_dir}/Prix{year}S1.csv")
            price_files.append(f"{data_dir}/Prix{year}S2.csv")
        else:
            price_files.append(f"{data_dir}/Prix{year}.csv")
    
    existing_files = [f for f in price_files if os.path.exists(f)]
    if existing_files:
        print(f"\n Found {len(existing_files)} existing price files")
        downloaded = {
            'prices': existing_files,
            'stations': f"{data_dir}/Stations2024.csv" if os.path.exists(f"{data_dir}/Stations2024.csv") else None,
            'services': f"{data_dir}/Services2024.csv" if os.path.exists(f"{data_dir}/Services2024.csv") else None
        }
    else:
        downloaded = {'prices': [], 'stations': None, 'services': None}
        print("\n No data files found. Please download manually and re-run.")
```

#### Define schema for gas price data

```python
# ============================================================================
# Define schema for gas price data
# ============================================================================

# Based on the README, fields are:
# id_pdv; cp; pop; latitude; longitude; date; id_carburant; nom_carburant; prix as define in at the begining in the project

# True means it is allowed to contain NULL values 
price_schema = StructType([
    StructField("station_id", LongType(), True),      # id pdv (point of sales)
    StructField("zip_code", StringType(), True),      # cp (zip code)
    StructField("pop_type", StringType(), True),      # pop (R: rural, A: urban)
    StructField("latitude_raw", LongType(), True),    # latitude (needs division)
    StructField("longitude_raw", LongType(), True),   # longitude (needs division)
    StructField("date_str", StringType(), True),      # date
    StructField("gas_id", IntegerType(), True),       # id carburant
    StructField("gas_name", StringType(), True),      # nom carburant
    StructField("price", DoubleType(), True)         # prix 
])

print("Schema defined for gas price data")
```

### Split date into year, month, week of year and Prepare latitude & longitude (divide by appropriate power of 10)

```python
# ============================================================================
# Split date into year, month, week of year
# ============================================================================

print("Applying transformations...")

# Transform the data
df_prices = df_prices.withColumn(
    # Convert date string to date type
    "date", F.to_date(F.col("date_str"), "yyyy-MM-dd")
).withColumn(
    # Extract year
    "year", F.year("date")
).withColumn(
    # Extract month
    "month", F.month("date")
).withColumn(
    # Extract week of year
    "week_of_year", F.weekofyear("date")
).withColumn(
    # Convert latitude (divide by 10^5 based on coordinates format)
    "latitude", F.col("latitude_raw") / 100000.0
).withColumn(
    # Convert longitude (divide by 10^5)
    "longitude", F.col("longitude_raw") / 100000.0
).withColumn(
    # Convert price from millieuros to euros
    "price_euros", F.col("price")  #/ 1000.0 # since price is alreading in Euros
).withColumn(
    # Extract department from zip code (first 2 digits)
    "department", F.substring("zip_code", 1, 2)   # This will be usefull later on to create maps
)

# Drop intermediate columns
df_prices = df_prices.drop("latitude_raw", "longitude_raw", "date_str")

print(" Transformations applied")
print("\nSchema after transformation:")
df_prices.printSchema()
```

#### Analyze gas types to determine which to exclude (2)

I am looking for the two gas types with the lowest number of records

```python
# ============================================================================
# Analyze gas types to determine which to exclude
# ============================================================================

print("=" * 60)
print("GAS TYPE ANALYSIS")
print("=" * 60)

# Count records and calculate statistics by gas type
gas_stats = spark.sql("""
    SELECT 
        gas_id,
        gas_name,
        COUNT(*) as record_count,
        COUNT(DISTINCT station_id) as station_count,
        ROUND(AVG(price_euros), 3) as avg_price,
        MIN(date) as first_date,
        MAX(date) as last_date
    FROM gas_prices
    GROUP BY gas_id, gas_name
    ORDER BY record_count DESC
""")

print("\nGas Types Statistics:")
gas_stats.show(truncate=False)

# Convert to pandas for easier analysis
gas_stats_pd = gas_stats.toPandas()

# Identify gas types with significantly fewer records
total_records = gas_stats_pd['record_count'].sum()
gas_stats_pd['percentage'] = (gas_stats_pd['record_count'] / total_records * 100).round(2)

print("\nPercentage of records per gas type:\n")
print(gas_stats_pd[['gas_name', 'record_count', 'percentage']].to_string(index=False))
```

**Remark**

In the official documenation, there are 6 gas types : Gazole, E10,SP98 ,E85, SP95, GPLc  ; 

None is not part of it and will be automatically removed.

**Gas type to be excluded :** GPLc and SP95 which contain 4 % and 6% of the data respectively and limited station availability, making analysis less meaningful

Code to automatically exclude the gas type

```python
# ============================================================================
# Code to automatically exclude the gas type
# ============================================================================

# Based on the statistics, I identify gas types with:
# - Very few records (< 1% of total) The None dataset
# - Limited station coverage
# - These are typically GPLc and E95 (Ethanol) in France

# Find the two gas types with the lowest record counts
gas_types_to_exclude = gas_stats_pd.nsmallest(3, 'record_count')['gas_name'].tolist()

print("=" * 60)
print("DECISION: Gas types to exclude")
print("=" * 60)
print(f"\nExcluding: {gas_types_to_exclude}")

# Update config
config['exclude_gas_types'] = gas_types_to_exclude

to_keep = list(set(  gas_stats_pd['gas_name'].unique() ) - set(gas_types_to_exclude))

config['to_keep_gas_types'] = to_keep 
```

#### Compute Price Index for each station per day

**Price Index Formula**

$$
\text{Price Index} = 100 \times \frac{\text{Day Price in Station} - \text{Average Day Price in France}}{\text{Average Day Price in France}} + 1
$$


```python
# ============================================================================
# COMPUTE DAILY AVERAGE PRICE IN FRANCE (per gas type)
# ============================================================================
# Calculate daily average price per gas type across all stations in France
daily_avg = df_prices_filtered.groupBy("date", "gas_name").agg(
    F.avg("price_euros").alias("avg_day_price_france")
)

# Join back to main dataframe
df_with_avg = df_prices_filtered.join(
    daily_avg,
    on=["date", "gas_name"],
    how="left"
)

df_with_avg.show(5)
print("\n Daily average prices computed")
```

```python
# ============================================================================
# Compute Price Index
# ============================================================================
# Price Index formula = 100 * ((Day Price in station - Average Day Price in France) / Average Day Price in France) + 1

df_with_index = df_with_avg.withColumn(
    "price_index",
    F.round(
        100 * (
            (F.col("price_euros") - F.col("avg_day_price_france")) / 
            F.col("avg_day_price_france")
        ) + 1,
        4
    )
)
df_with_index.show(5)


print("Price Index computed")
print("\nPrice Index interpretation (with the +1 it makes it tricky):")
print("  - Index = 1: Station price equals national average")
print("  - Index > 1: Station price above national average")
print("  - Index < 1: Station price below national average")
print("  - The magnitude of (Index − 1) / 100 represents the relative price difference")
```

For example for index = -1.0333

```python
first_index = -1.0333
indexval = (first_index - 1) / 100
print(indexval)
```

This means station is about **2%** cheaper

#### Compute Week Index (weeks since first week)

This is a continuous time counter that measures how many weeks have passed since the first date in the dataset. 

In orther words, week index is the difference between the first date in the dataset and all other dates, expressed in weeks.

```python
# ============================================================================
# Compute Week Index
# ============================================================================

# Get the first date in the dataset
first_date = df_with_index.agg(F.min("date")).collect()[0][0]
print(f"First date in dataset: {first_date}")

# Calculate week index: number of weeks since the first week
# I use datediff to get days difference, then divide by 7

df_final = df_with_index.withColumn(
    "week_index",
    F.floor(F.datediff(F.col("date"), F.lit(first_date)) / 7) + 1
)

# Verify week index
print("\nWeek Index verification:")
df_final.select("date", "year", "week_of_year", "week_index") \
    .orderBy("date") \
    .distinct() \
    .show(10)

# Check max week index
max_week = df_final.agg(F.max("week_index")).collect()[0][0]
print(f"Maximum week index: {max_week}")
print(f"Expected for 3 years: ~156 weeks")
```

- In the data, the corse department code is 20 but in the department geojson file it is given by ["2B", "2B"]
- I haD to normalise the code so to assign the right department to the right index price

```python
# ============================================================================
# CREATE FOLIUM MAP WITH HOVER TOOLTIPS
# ============================================================================
def create_map(data=france_geo, gas_type_to_map="SP98", output_name="france_gas_prices_map.html"):
    
    # Select gas type
    map_data = dept_price_index[dept_price_index["gas_name"] == gas_type_to_map].copy()

    # Ensure department codes are strings
    map_data["department"] = map_data["department"].astype(str)

    # Create lookup dictionary
    dept_to_index = dict(zip(map_data["department"], map_data["avg_price_index"]))

    # Create map
    m = folium.Map(location=[46.603354, 1.888334], zoom_start=6)

    # Normalize GeoJSON department codes (2A, 2B -> 20)
    for feature in data["features"]:
        code = feature["properties"]["code"]
        if code in ["2A", "2B"]:
            feature["properties"]["code"] = "20"

    # Create choropleth
    choropleth = folium.Choropleth(
        geo_data=data,
        name="Price Index",
        data=map_data,
        columns=["department", "avg_price_index"],
        key_on="feature.properties.code",
        fill_color="RdYlGn_r",
        fill_opacity=0.7,
        line_opacity=1,
        legend_name=f"{gas_type_to_map} Price Index (1 = National Average)"
    ).add_to(m)


    # INCREASE COLOR BAR SIZE
    # Access the color map and modify its properties
    for key in choropleth._children:
        if key.startswith('color_map'):
            color_map = choropleth._children[key]
            # Modify width and height
            color_map.width = 450   # Default is ~300, increase for wider
            color_map.height = 45   # Default is ~40, increase for taller
            break

    
    # Inject tooltip data
    for feature in choropleth.geojson.data["features"]:
        dept_code = feature["properties"]["code"]
        price_idx = dept_to_index.get(dept_code)
        feature["properties"]["price_index"] = (
            f"{price_idx:.4f}" if price_idx is not None else "N/A"
        )

    # Tooltip
    folium.GeoJsonTooltip(
        fields=["nom", "code", "price_index"],
        aliases=["Department:", "Code:", "Price Index:"],
        localize=True,
        sticky=False,
        style="""
            background-color: white;
            border: 2px solid black;
            border-radius: 3px;
            box-shadow: 3px 3px 3px rgba(0,0,0,0.3);
            font-size: 12px;
            padding: 8px;
        """
    ).add_to(choropleth.geojson)

    folium.LayerControl().add_to(m)

    m.save(output_name)
    print(f" Map saved as {output_name}")
    print(f"  Gas type: {gas_type_to_map}")

    return m
```

## 5. Modeling - Price Forecasting

Build a model to forecast the next day's gas price using Spark ML for a gas type in a station

### Approach:
- Use lag features (previous day prices)
- Include temporal features (day of week, month)
- Compare Linear Regression vs Random Forest

reference link: https://arxiv.org/abs/2101.02118


**Preparing features fro modelling**

From the previous analysis, Gazole is the gas type with the highest number of records, representing **29%** of the dataset. Therefore, it will be used for the forecasting task, although other gas types can also be analyzed by updating the variable **gas_type_model** below.

```python
# ============================================================================
# Preparing features fro modelling
# ============================================================================
print("Preparing modeling features...")

# Select a specific gas type for modeling (e.g., SP95 or Gazole, E85.. )
gas_type_model = "Gazole"  # Most common gas type in France

# Filter for one gas type
df_model = df_final.filter(F.col("gas_name") == gas_type_model)

print(f"Modeling gas type: {gas_type_model}")
print(f"Records for modeling: {df_model.count():,}")
```

```python
# ============================================================================
# CREATE LAG FEATURES
# ============================================================================

# Define window for lag features (partition by station, order by date)
window_spec = Window.partitionBy("station_id").orderBy("date")  # each station is treated independently

# Create lag features
df_model_lag = df_model \
    .withColumn("price_lag_1", F.lag("price_euros", 1).over(window_spec)) \
    .withColumn("price_lag_2", F.lag("price_euros", 2).over(window_spec)) \
    .withColumn("price_lag_3", F.lag("price_euros", 3).over(window_spec)) \
    .withColumn("price_lag_7", F.lag("price_euros", 7).over(window_spec)) \
    .withColumn("day_of_week", F.dayofweek("date")) \
    .withColumn("day_of_month", F.dayofmonth("date"))

# Calculate rolling average (7-day)
window_7d = Window.partitionBy("station_id").orderBy("date").rowsBetween(-6, 0)
df_model_lag = df_model_lag.withColumn(
    "price_rolling_7d", 
    F.avg("price_euros").over(window_7d)
)

# Remove rows with null lag values
df_model_clean = df_model_lag.dropna(subset=[
    "price_lag_1", "price_lag_2", "price_lag_3", "price_lag_7", "price_rolling_7d"
])

print(f"\nRecords after creating lag features: {df_model_clean.count():,}")
print("\nFeatures created:")
print("  - price_lag_1: Price 1 day ago")
print("  - price_lag_2: Price 2 days ago")
print("  - price_lag_3: Price 3 days ago")
print("  - price_lag_7: Price 7 days ago")
print("  - price_rolling_7d: 7-day rolling average : average of the last 7 days")
print("  - day_of_week: Day of week (1-7)")
print("  - day_of_month: Day of month (1-31)")
print("  - month: Month of year (1-12)")
```

- The lag features were selected to capture the most relevant short-term temporal dependencies in gas price dynamics while maintaining a simple and robust modeling pipeline.

- Intermediate lags are usually highly correlated with nearby lags and the rolling average, so removing them reduces multicollinearity, limits overfitting, and improves model robustness while keeping the pipeline simple.

- The difference between consecutive days is usually small compared to the price level.

```python
# ============================================================================
# PREPARE ML PIPELINE
# ============================================================================

# Define feature columns
feature_cols = [
    "price_lag_1", "price_lag_2", "price_lag_3", "price_lag_7",
    "price_rolling_7d", "day_of_week", "day_of_month", "month"
]

# Vector assembler to combine features
assembler = VectorAssembler(
    inputCols=feature_cols,
    outputCol="features_raw"
)

# Standard scaler
scaler = StandardScaler(
    inputCol="features_raw",
    outputCol="features",
    withStd=True,
    withMean=True
)

print(" Feature assembler and scaler defined")
print(f"\nFeature columns: {feature_cols}")
```

```python
df_model_clean[['date']+feature_cols].show(5)
```

```python
# ============================================================================
# TRAIN-TEST SPLIT (Time-based)
# ============================================================================

# For time series, I use a temporal split instead of random
# Train on earlier data, test on later data

# Find split date (use last 2 months for testing)
max_date = df_model_clean.agg(F.max("date")).collect()[0][0]
split_date = max_date - pd.Timedelta(days=60)  # ~2 months for testing

print(f"Max date in data: {max_date}")
print(f"Train/Test split date: {split_date}")

# Split data
train_data = df_model_clean.filter(F.col("date") < F.lit(split_date))
test_data = df_model_clean.filter(F.col("date") >= F.lit(split_date))

print(f"\nTraining records: {train_data.count():,}")
print(f"Testing records: {test_data.count():,}")
```

```python
train_data.show(2)
```

```python
# ============================================================================
# MODEL 1: LINEAR REGRESSION
# ============================================================================

print("=" * 60)
print("MODEL 1: LINEAR REGRESSION")
print("=" * 60)

# Define Linear Regression model
lr = LinearRegression(
    featuresCol="features",
    labelCol="price_euros",
    predictionCol="prediction",
    maxIter=100,
    regParam=0.01,
    elasticNetParam=0.5
)

# Build pipeline
lr_pipeline = Pipeline(stages=[assembler, scaler, lr])

# Train model
print("\nTraining Linear Regression model...")
lr_model = lr_pipeline.fit(train_data)
print(" Training complete")

# Make predictions
lr_predictions = lr_model.transform(test_data)
```

```python
lr_model.transform(test_data).show(5)
```

```python
# ============================================================================
# MODEL 2: RANDOM FOREST
# ============================================================================

print("=" * 60)
print("MODEL 2: RANDOM FOREST")
print("=" * 60)

# Define Random Forest model
rf = RandomForestRegressor(
    featuresCol="features",
    labelCol="price_euros",
    predictionCol="prediction",
    numTrees=50,
    maxDepth=10,
    seed=42
)

# Build pipeline
rf_pipeline = Pipeline(stages=[assembler, scaler, rf])

# Train model
print("\nTraining Random Forest model...")
rf_model = rf_pipeline.fit(train_data)
print(" Training complete")

# Make predictions
rf_predictions = rf_model.transform(test_data)
```

```python
# ============================================================================
# MODEL EVALUATION
# ============================================================================

print("=" * 60)
print("MODEL EVALUATION")
print("=" * 60)

# Define evaluators
evaluators = {
    'RMSE': RegressionEvaluator(labelCol="price_euros", predictionCol="prediction", metricName="rmse"),
    'MAE': RegressionEvaluator(labelCol="price_euros", predictionCol="prediction", metricName="mae"),
    'R2': RegressionEvaluator(labelCol="price_euros", predictionCol="prediction", metricName="r2")
}

# Evaluate both models
results = []
for model_name, predictions in [('Linear Regression', lr_predictions), ('Random Forest', rf_predictions)]:
    metrics = {'Model': model_name}
    for metric_name, evaluator in evaluators.items():
        metrics[metric_name] = round(evaluator.evaluate(predictions), 4)
    results.append(metrics)

# Display results
results_df = pd.DataFrame(results)
print("\nModel Performance Comparison:")
print(results_df.to_string(index=False))

print("\nMetrics Interpretation:")
print("  - RMSE: Root Mean Squared Error (lower is better)")
print("  - MAE: Mean Absolute Error (lower is better)")
print("  - R2: Coefficient of Determination (higher is better, max=1)")
```
